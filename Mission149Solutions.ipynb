```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
cb_gray = (171/255, 171/255, 171/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture', 'Business', 'Architecture']



def add_column_of_charts(fig, data, s, cats, first_lab, last_lab):
    starting_pos = s
    for sp in range(0,len(cats)):
        ax = fig.add_subplot(6,3,starting_pos)
        ax.plot(women_degrees['Year'], women_degrees[cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
        ax.plot(women_degrees['Year'], 100-women_degrees[cats[sp]], c=cb_orange, label='Men', linewidth=3)
        ax.set_title(cats[sp])
        for spine in ax.spines:   
            ax.spines[spine].set_visible(False)    
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.set_yticks([0,100])
        ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
        ax.axhline(50, c=cb_gray, alpha=0.5)
        
        if sp == 0:
            ax.text(2005, first_lab[0], 'Men')
            ax.text(2005, first_lab[1], 'Women')
        elif sp == (len(cats)-1):
            ax.text(2005, last_lab[0], 'Men')
            ax.text(2005, last_lab[1], 'Women')
            ax.tick_params(labelbottom='on')
        starting_pos += 3

fig = plt.figure(figsize=(18, 18))
add_column_of_charts(fig, women_degrees, 1, stem_cats, [87, 8], [62, 35])
add_column_of_charts(fig, women_degrees, 2, lib_arts_cats, [75, 22], [57, 39])
add_column_of_charts(fig, women_degrees, 3, other_cats, [90, 7], [62, 35])

plt.show()
```
